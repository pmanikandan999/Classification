```python
#from sklearn.datasets.samples_generator import make_blobs
supervised, clusters = make_blobs(n_samples = 600, centers = 8, cluster_std = 0.75, random_state = 0)
 
# Creating the BIRCH clustering model
model = Birch(branching_factor = 50, n_clusters = None, threshold = 1.5)
 
# Fit the data (Training)
model.fit(supervised)
 
# Predict the same data
pred = model.predict(supervised)
 
# Creating a scatter plot
plt.scatter(supervised[:, 0], supervised[:, 1], c = pred, cmap = 'rainbow', alpha = 0.7, edgecolors = 'b')
plt.show()
```
